```python
import os
import requests
from tqdm import tqdm
```

```python
SYNSETS = {
    "bus": ["n02892201"],
    
    "truck": [
        "n04467665",  # trailer truck
        "n04467332",  # pickup truck
        "n08641617"   # tow truck
    ],
    
    "car": [
        "n02958343",  # car
        "n03100232",  # convertible
        "n03594945",  # jeep
        "n03769881",  # minivan
        "n04037443",  # race car
        "n04285008"   # sports car
    ],
    
    "bike": [
        "n03790512",  # motorcycle
        "n02835271"   # bicycle
    ]
}
BASE_URL = "https://image-net.org/api/text/imagenet.synset.geturls?wnid="
```

```python
def download_class(class_name, synsets):

    os.makedirs(class_name, exist_ok=True)
    img_count = 0

    for synset in synsets:

        print(f"\nDownloading synset {synset}")

        url = BASE_URL + synset
        response = requests.get(url)

        urls = response.text.split("\n")

        for link in tqdm(urls):

            # skip empty links
            if link.strip() == "":
                continue

            try:
                img = requests.get(link, timeout=5)

                # check if request worked
                if img.status_code != 200:
                    continue

                # check if it is an image
                if "image" not in img.headers.get("Content-Type",""):
                    continue

                file_path = os.path.join(
                    class_name,
                    f"{class_name}_{img_count}.jpg"
                )

                with open(file_path, "wb") as f:
                    f.write(img.content)

                img_count += 1

            except:
                continue
```

```python
def main():

    for class_name, synsets in SYNSETS.items():

        print("\n=================================")
        print(f"Downloading class: {class_name}")
        print("=================================")

        download_class(class_name, synsets)
```

```python

if __name__ == "__main__":
    main()
```
